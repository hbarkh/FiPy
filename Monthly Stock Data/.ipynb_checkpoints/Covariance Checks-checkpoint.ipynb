```python
import pandas as pd
import numpy as np
import edhec_risk_kit as erk
%config Completer.use_jedi = False
```

```python
HSU = pd.read_csv("HSU.TO.csv", index_col='Date', parse_dates=True)[["Close"]]
HSU.columns = ['HSU']

SOXL = pd.read_csv("HSU.TO.csv", index_col='Date', parse_dates=True)[["Close"]]
SOXL.columns = ['SOXL']

HQU = pd.read_csv("HQU.TO.csv", index_col='Date', parse_dates=True)[["Close"]]
HQU.columns = ['HQU']

ZST = pd.read_csv("ZST.TO.csv", index_col='Date', parse_dates=True)[["Close"]]
ZST.columns = ['ZST']

#stocks = pd.concat([HQU,HSU], axis = 1, ignore_index=True)
#stocks
print(type(HSU))

df = pd.concat([HQU,HSU,SOXL],axis=1)
df = df.pct_change().dropna()
cov = df.cov()
correlation = df.cov()/(np.prod(df.std().values))
```

```python
rets = erk.annualize_rets(df, 12)
rets
```
