### Rolling Correlations
along with MultiIndexes and '.grouby'

```python
ts_corr = ind_returns.rolling(window=36).corr()
ts_corr.tail()

# notice how it has two indices. Meaning that in 2018-12 the correlations were such:
```

```python
ind_returns.head()
```

what we want is to look at the average correlation of everything in the market with each other, seen by date
So we want to see whether correlations hold during market downturns hint: they dont! markets go down in lockstep
Therefore, lets use pandas group by

```python
ts_corr.groupby(level=0) #alternatively, can level the index ts_corr.index.names = ['date','industry']
ind_tr36corr = ts_corr.groupby(level=0).apply(lambda cormat: cormat.values.mean())
```

```python
ind_tr36corr.plot()
```

```python
tmi_tr36rets.plot(label = "Trailing 36 months returns", secondary_y=True, color = 'orange')
ind_tr36corr.plot(label = "Trailing 36 months correlations")

plt.show()
```

```python
tmi_tr36rets.corr(ind_tr36corr) #the correlation of average correlation of industries and returns is negative!
# as markets fall, correlations increase. Diversification doesn't help against systemic risk.
```
