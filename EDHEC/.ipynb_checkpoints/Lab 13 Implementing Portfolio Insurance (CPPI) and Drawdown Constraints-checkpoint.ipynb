```python
risky_r = ind_return["2000":][["Steel", "Fin", "Beer"]]
risky_r.shape
```

```python
# gives dataframe of same shape as existing
safe_r = pd.DataFrame().reindex_like(risky_r)
safe_r.shape
```

```python
safe_r[:] = 0.02/12
start = 1000
floor = 0.8
```

### Steps in the CPPI algorithm

1. Cushion - Asset value minus Floor value
2. Compute an allocation to safe and rsiky assets with: m * (risk budget)
3. Recompute the asset value based on the returns

```python
dates = risky_r.index
n_steps = len(dates)
account_value = start
floor_value = start * floor
m = 3

# for backtesting:
account_history = pd.DataFrame().reindex_like(risky_r)
cushion_history = pd.DataFrame().reindex_like(risky_r)
risky_w_history = pd.DataFrame().reindex_like(risky_r)

for step in range(n_steps):
    
    cushion = (account_value - floor_value)/account_value
    
    risky_w = m * cushion    
    risky_w = np.minimum(risky_w, 1)  # no leverage
    risky_w = np.maximum(risky_w, 0)  # no shorting
    
    safe_w = 1 - risky_w
    
    risky_alloc = account_value*risky_w
    safe_alloc = account_value*safe_w
    
    # Update the account value for this time step
    account_value = risky_alloc*(1 + risky_r.iloc[step]) + safe_alloc*(1 + safe_r.iloc[step])
    
    # Save the values to plot and analyze
    account_history.iloc[step] = account_value
    cushion_history.iloc[step] = cushion
    risky_w_history.iloc[step] = risky_w
```

```python
account_history.plot()
```

```python
risky_wealth = start * (1+risky_r).cumprod()
```

```python
risky_wealth.plot()
```

```python
ax = account_history["Beer"].plot(label = "CPPI")
ax = risky_wealth["Beer"].plot(label = "Risky")
ax.axhline(floor_value,color ="red", linestyle = '--')

plt.legend()
plt.show()
```

```python
risky_w_history.plot(title='CPPI Weight Allocations Over Time')
```

```python
ax = account_history["Fin"].plot(label = "CPPI")
ax = risky_wealth["Fin"].plot(label = "Risky")
ax.axhline(floor_value,color ="red", linestyle = '--')

plt.legend()
plt.show()
```

```python
ax = account_history["Steel"].plot(label = "CPPI")
ax = risky_wealth["Steel"].plot(label = "Risky")
ax.axhline(floor_value,color ="red", linestyle = '--',label='floor')

plt.legend()
plt.show()
```
